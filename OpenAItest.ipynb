```python
import openai
```

```python
OPENAI_API_KEY
openai.api_key = OPENAI_API_KEY
model = "gpt-3.5-turbo-0125"
```

```python
query = "프랑스의 수도를 알려줘"

messages = [{
    "role": "user",
    "content": query
}]
```

```python
response = openai.ChatCompletion.create(
    model=model,
    messages=messages
)

answer = response.choices[0].message.content
print(answer)
```
